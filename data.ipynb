```python
import openpyxl 

DTP_RAW_TABLAS = 'data/DTP_RAW/DTP-TABLAS-2023.xlsx'

dtp = pd.read_excel(DTP_RAW_TABLAS, sheet_name='T7.1', skiprows=2, usecols='B:E', nrows=13)
dtp.head(15)
```

```python
#make first row the header
dtp.columns = dtp.iloc[0]
dtp.drop(0, inplace=True)
```

```python
#change column names
dtp.columns = ['Año', 'Ingresos Operacionales', 'Ingresos subsidio', 'Egresos']
dtp.info()
```

```python
dtp['Año'] = dtp['Año'].astype(int)
dtp 
```

```python
def to_pesos(x):
    x = x.strip('$')
    x = x.replace('.', '')
    return int(x)

for col in dtp.columns[1:]:
    dtp[col] = dtp[col].apply(to_pesos)

dtp.info()
```

```python
dtp 
```

```python
#save csv 
dtp.to_csv('data/DTP_CSV/TABLA-INGRESOS-EN-MILLONES.csv', index=False)
```

```python

plt.plot(dtp['Año'], dtp['Ingresos Operacionales'])
plt.plot(dtp['Año'], dtp['Ingresos subsidio'])
plt.plot(dtp['Año'], dtp['Egresos'])
plt.xticks(dtp['Año'], rotation=45)
plt.legend(['Ingresos Operacionales', 'Ingresos subsidio', 'Egresos'])
plt.show()
```

```python
import pandas as pd 
pobreza_2020 = pd.read_csv('data/SOCIO-ECONOMICA/POBREZA_2020.csv', sep=';')
pobreza_2020.head()
```

```python
#RESET INDEX
pobreza_2020.reset_index(drop=True, inplace=True)
```

```python
#first row as header
pobreza_2020.columns = pobreza_2020.iloc[0]
poverty_2020 = pobreza_2020[1:]
poverty_2020.head()
```

```python
pobreza_2020 = pobreza_2020[pobreza_2020['Región'] == 'XIII Metropolitana de Santiago']
pobreza_2020.head()
```

```python
pobreza_2020.info()
```

```python
def transformar_float(x):
    x = x.replace('%', '')
    x= x.replace(',', '.')
    return float(x)

def transformar_int(x):
    x = x.replace('.', '')
    return int(x)
```

```python
pobreza_2020['Número de personas en situación de pobreza por ingresos (**)'] = pobreza_2020['Número de personas en situación de pobreza por ingresos (**)'].apply(transformar_int)
```

```python
pobreza_2020['Número de personas según proyecciones de población (*)'] = pobreza_2020['Número de personas según proyecciones de población (*)'].apply(transformar_int)
```

```python
pobreza_2020['Porcentaje de personas en situación de pobreza por ingresos 2020'] = pobreza_2020['Porcentaje de personas en situación de pobreza por ingresos 2020'].apply(transformar_float)
pobreza_2020['Límite superior'] = pobreza_2020['Límite superior'].apply(transformar_float)
pobreza_2020['Límite inferior'] = pobreza_2020['Límite inferior'].apply(transformar_float)
```

```python
pobreza_2020.info()
```

```python
pobreza_2020.to_csv('data/SE-FILTRADA/POBREZA-METROPOLITANA-2020.csv', index=False)
```
