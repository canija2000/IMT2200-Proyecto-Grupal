```python
valor_historico_tarifas = pd.read_excel('data/DTP_RAW/DTP-ANEXOS-2023.xlsx', sheet_name='A16', skiprows=3, usecols='B:I', nrows=204)
```

```python
valor_historico_tarifas.head(200
                             )
```

```python
valor_historico_tarifas.info()
```

```python
for col in valor_historico_tarifas.columns[1:]:
    valor_historico_tarifas[col] = valor_historico_tarifas[col].fillna(0)
    valor_historico_tarifas[col] = valor_historico_tarifas[col].astype(str)
    valor_historico_tarifas[col]= valor_historico_tarifas[col].apply(lambda x : 0 if x == 'nan' else x)
    valor_historico_tarifas[col] = valor_historico_tarifas[col].apply(lambda x: x.replace('$','').strip())
    valor_historico_tarifas[col] = valor_historico_tarifas[col].astype(int)

valor_historico_tarifas.info()
```

```python
valor_historico_tarifas.to_csv('data/DTP_CSV/tabla-valor_historico_tarifas.csv', index=False)
```

```python
evasion = pd.read_excel('data/DTP_RAW/DTP-ANEXOS-2023.xlsx', sheet_name='A17', skiprows=4, usecols='B:N', nrows=17)


evasion.head(22)
```

```python
evasion.info()
```

```python
def clean_evasion(df):
    for col in df.columns[1:]:
        df[col] = df[col].fillna(0)
        df[col] = df[col].astype(str)
        df[col] = df[col].apply(lambda x: 0 if x in 'sin medición' else x)
        df[col] = df[col].astype(float)
    return df

evasion = clean_evasion(evasion)

evasion.info()
```

```python
evasion.to_csv('data/DTP_CSV/tabla-evasion-historica.csv', index=False)
```
