```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from os.path import join
```

```python
hogares_edad = pd.read_csv(
    join("data", "hogares_segun_tramo_CSE_por_edad", "ind_6_105_201606.csv"),
    sep=";",
    skiprows=2
)
hogares_edad
```

```python
hogares_edad.dtypes.to_dict()
```

```python
# Las columnas tienen valores inesperados en forma de string
hogares_edad[hogares_edad["unidad vecinal"].str.isnumeric() == False]
```

```python
# Datos de RM
hogares_edad = hogares_edad[hogares_edad["región"] == 13]
```

```python
hogares_edad.columns.to_list()
```

```python
def is_uv(col_name):
    if col_name[-2:] == "uv":
        return True
    else:
        return False
    
# Quitamos los datos de unidades vecinales
cols_keep = [col for col in hogares_edad.columns.to_list() if not is_uv(col)]
cols_keep.remove("unidad vecinal")

hogares_edad = hogares_edad[cols_keep]
```

```python
def is_year(col_name):
    if "años" in col_name:
        return True
    else:
        return False

# Quitamos los datos de rangos etarios 
cols_keep = [col for col in hogares_edad.columns.to_list() if not is_year(col)]
hogares_edad = hogares_edad[cols_keep]
```

```python
def is_com(text):
    if text[-6:] == "comuna":
        return True
    else:
        return False
    

# Mantenemos solo la data de comunas
cols_keep = [col for col in hogares_edad.columns if is_com(col)]
hogares_edad = hogares_edad[cols_keep]
```

```python
# Eliminamos datos duplicados
hogares_edad.drop_duplicates(inplace=True, ignore_index=True)
hogares_edad
```

```python
hogares_edad.columns = [
    "codigo comuna",
    "hogares tramo 0 - 40",
    "hogares tramo 41 - 50",
    "hogares tramo 51 - 60",
    "hogares tramo 61 - 70",
    "hogares tramo 71 - 80",
    "hogares tramo 81 - 90",
    "hogares tramo 91 - 100",
    "total hogares"
]
hogares_edad
```

```python
codigos_comunales = pd.read_csv(join("data", "codigos_comunales.csv"))
codigos_comunales
```

```python
hogares_edad = hogares_edad.astype(int)
df_final = pd.merge(hogares_edad, codigos_comunales, how="right", on="codigo comuna")
df_final
```

```python
df_pobreza = df_final[["codigo comuna", "nombre comuna", "hogares tramo 0 - 40"]]
df_pobreza["prc tramo 0 - 40"] = df_final["hogares tramo 0 - 40"] / df_final["total hogares"]
df_pobreza.sort_values("prc tramo 0 - 40", inplace=True, ignore_index=True)

df_pobreza
```

```python
df_riqueza = df_final[["codigo comuna", "nombre comuna", "hogares tramo 71 - 80", "hogares tramo 81 - 90", "hogares tramo 91 - 100"]]
df_riqueza["prc tramos altos"] = (df_final["hogares tramo 71 - 80"] + \
                                  df_final["hogares tramo 81 - 90"] + \
                                  df_final["hogares tramo 91 - 100"]) / df_final["total hogares"]

df_riqueza.sort_values("prc tramos altos", inplace=True, ignore_index=True)
df_riqueza
```

```python
import geopandas as gpd

mapa: gpd.GeoDataFrame = gpd.read_file(join("data", "r13.geojson"))
mapa["codigo_comuna"] = mapa["codigo_comuna"].astype(int)
mapa
```

```python
mapa = mapa.dissolve("codigo_comuna")
mapa
```

```python
mapa = pd.merge(mapa, df_pobreza[["codigo comuna", "prc tramo 0 - 40"]], how="left", left_index=True, right_on="codigo comuna")
mapa = pd.merge(mapa, df_riqueza[["codigo comuna", "prc tramos altos", "nombre comuna"]], how="left", on="codigo comuna")
mapa
```

```python
mapa.plot(
    column = "prc tramo 0 - 40",
    cmap = "Reds",
    legend = True,
    figsize = (12, 12)
).set_title("Porcentaje de hogares en tramo CSE 0% - 40%")
```

```python
mapa.plot(
    column = "prc tramos altos",
    cmap = "Greens",
    legend = True,
    figsize = (12, 12)
).set_title("Porcentaje de hogares en tramo CSE 70% - 100%")
```
