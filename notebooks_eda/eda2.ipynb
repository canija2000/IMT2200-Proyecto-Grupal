```python
tp = gpd.read_file('../Zonas Transporte Pblico/Cumunas- Zonas.shp')
```

```python
tp.plot()
```

```python
tp.head()
```

```python
tp['ZONA_TS'].unique()
```

```python
tp['ZONA_TS'] = tp['ZONA_TS'].apply(lambda x:  'Zona M' if x== None else x)
```

```python
import matplotlib.pyplot as plt

colors = [
    "#4E79A7",  
    "#F28E2B",  
    "#E15759",  
    "#76B7B2",  
    "#59A14F",  
    "#EDC948",  
    "#B07AA1",  
    "#FF9DA7",  
    "#9C755F",
    "#BAB0AC"   
]

unique_zones = tp['ZONA_TS'].unique()
if len(colors) < len(unique_zones):
    raise ValueError("No hay suficientes colores para asignar a todas las zonas.")


zone_color_mapping = {zone: color for zone, color in zip(unique_zones, colors)}


tp['color'] = tp['ZONA_TS'].map(zone_color_mapping)


fig, ax = plt.subplots(1, 1, figsize=(10, 10))
tp.plot(color=tp['color'], ax=ax)


handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=zone) 
           for zone, color in zone_color_mapping.items()]
ax.legend(handles=handles, title="Zonas TS", loc='center left', bbox_to_anchor=(1, 0.5))

# poner nombres a las comunas

for x, y, label in zip(tp.geometry.centroid.x, tp.geometry.centroid.y, tp['COMUNA']):
    ax.text(x, y, label, fontsize=6, ha='center')


plt.title('Zonas de Transporte Pblico')
plt.show()
fig.savefig('../imgs/Zonas de Transporte Pblico.png', dpi=300, bbox_inches='tight')
```

```python
num_buses = pd.read_csv("../data/DTP_CSV/TABLA-NUMERO-BUSES-POR-AO.csv")
```

```python
num_buses.head()
```

```python
anos = num_buses.columns[1:]
anos = [int(a) for a in anos]
anos 
```

```python
n_buses = num_buses.iloc[0, 1:]
n_buses = [int(n) for n in n_buses]
```

```python
conductores = pd.read_csv("../data/DTP_CSV/TABLA-CONDUCTORES.csv")
```

```python
conductores.head()
```

```python
import queue

a = [0,0]
cond = conductores.iloc[0, 1:]
cond = [int(c) for c in cond]
a += cond

a

```

```python
b= [0,0]
conda = conductores.iloc[1, 1:]
conda = [int(c) for c in conda]

b += conda
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
ax.plot(anos, n_buses, marker='o')
ax.plot(anos, a, marker='o', color='red', label='Conductores')
ax.plot(anos, b, marker='o', color='green', label='Conductoras')
ax.set_xlabel('Ao')
ax.set_ylabel('Nmero de buses')
ax.set_title('Nmero de buses por ao')
ax.grid(True)
ax.set_xticks(anos);
ax.legend()
fig.savefig('../imgs/Numero de buses por ao.png', dpi=300, bbox_inches='tight')
```
